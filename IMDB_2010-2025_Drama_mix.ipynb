```python
# Define the URL
url = "https://www.imdb.com/list/ls566941243/"
```

```python
# Set up Chrome options to run the browser in incognito mode
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--incognito")

# Initialize the Chrome driver with the specified options
driver = webdriver.Chrome(options=chrome_options)

# Navigate to the IMDb list URL
driver.get(url)

# Wait for the page to load (adjust the wait time according to your webpage)
driver.implicitly_wait(10)

# Get the HTML content of the page after it has fully loaded
html_content = driver.page_source

# Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
# Extract movie data
movies_data = []
for movie in soup.find_all('div', class_='sc-ad5a2436-0'):
    title_tag = movie.find('h3', class_='ipc-title__text')
    title = title_tag.text if title_tag else 'N/A'
    genre_tag = movie.find('span', class_='sc-ad5a2436-7 cJVQtZ dli-title-metadata-item')
    genre = genre_tag.text.strip() if genre_tag else 'N/A'
    rating_tag = movie.find('span', class_='ipc-rating-star--rating')
    rating = rating_tag.text if rating_tag else 'N/A'
    runtime_tag = movie.find('span', class_='sc-ad5a2436-7 cJVQtZ dli-title-metadata-item')
    runtime = runtime_tag.text if runtime_tag else 'N/A'
    stars = 'N/A'  # IMDb doesn't provide stars in the way your original script assumes
    
    movies_data.append([title, genre, stars, runtime, rating])
# Create a Pandas DataFrame from the collected movie data
df = pd.DataFrame(movies_data, columns=['Title', 'Genre', 'Stars', 'Runtime', 'Rating'])


# Close the Chrome driver
driver.quit()
```

```python
# Import the necessary modules
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
import time


# Define the Star class
class Star:
    def __init__(self, name):
        self.name = name

    def __repr__(self):
        return f"Star(name={self.name})"

# Initialize the WebDriver
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))

# Open the target page
driver.get("https://www.imdb.com/search/name/?birth_monthday=03-07")

# Wait for the page to load
time.sleep(5)

# Extract star names
star_elements = driver.find_elements(By.CSS_SELECTOR, ".ipc-title-link-wrapper")

# Parse the data
stars = []
for element in star_elements:
    name = element.text
    stars.append(Star(name))

# Print the list of celebrities born on 03-07
print(stars)

# Close the browser
driver.quit()
```

Recently, Google Chrome has introduced its own official headless mode (without the use of third-party libraries and plugins like Selenium WebDriver).

The native Chrome protocol, designed to manage the browser from third-party programs, is called the Chrome DevTools Protocol. Mind that it's supported not only by Chrome but also by many browsers built on its code base, including pure Chromium and a special docker build for parsing called headless-shell.

The main problem with integrating headless Chrome with any external software is the need to write a special connector that could convert programmatic calls into commands and syntax for the Chrome DevTools Protocol.

This is exactly the task that the Chromedp library addresses. It allows managing headless Chrome without involving external dependencies from scripts.

Naturally, Chromedp is most often used for parsing web pages and organizing testing (websites and web applications).

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time
import pandas as pd

# Define the Movie class
class Movie:
    def __init__(self, title, year, runtime, rating):
        self.title = title
        self.year = year
        self.runtime = runtime
        self.rating = rating

    def __repr__(self):
        return f"Movie(title={self.title}, year={self.year}, runtime={self.runtime}, rating={self.rating})"

# Initialize the WebDriver
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))

# Open the target page
driver.get("https://www.imdb.com/search/title/?title_type=feature&genres=crime,horror,mystery&release_date=2015-01-01,2025-01-31")

# Wait for the page to load
time.sleep(5)

# Extract movie details
movie_elements = driver.find_elements(By.CSS_SELECTOR, "li.ipc-metadata-list-summary-item")

# Parse the data
movies = []
for element in movie_elements:
    title = element.find_element(By.CSS_SELECTOR, "h3.ipc-title__text").text
    year = element.find_elements(By.CSS_SELECTOR, "span.sc-d5ea4b9d-7")[0].text
    runtime = element.find_elements(By.CSS_SELECTOR, "span.sc-d5ea4b9d-7")[1].text
    rating = element.find_element(By.CSS_SELECTOR, "span.ipc-rating-star--rating").text
    movies.append(Movie(title, year, runtime, rating))

# Create a DataFrame
df = pd.DataFrame([vars(movie) for movie in movies])

# Close the browser
driver.quit()

# Print the DataFrame
df
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time
import pandas as pd

# Define the Movie class
class Movie:
    def __init__(self, title, year, runtime, rating, mpa_rating):
        self.title = title
        self.year = year
        self.runtime = runtime
        self.rating = rating
        self.mpa_rating = mpa_rating

    def __repr__(self):
        return f"Movie(title={self.title}, year={self.year}, runtime={self.runtime}, rating={self.rating}, mpa_rating={self.mpa_rating})"

# Set Chrome options to include the Accept-Language header
chrome_options = Options()
chrome_options.add_argument("--lang=en-US")

# Initialize the WebDriver with options
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=chrome_options)

# Open the target page
base_url = "https://www.imdb.com/search/title/?title_type=feature&genres=crime,horror,mystery&release_date=2015-01-01,2025-01-31"
driver.get(base_url)

# Wait for the page to load
time.sleep(5)

# Initialize an empty list to store movies
movies = []

# Loop through 50 pages
for page in range(1, 51):
    # Extract movie details
    movie_elements = driver.find_elements(By.CSS_SELECTOR, "li.ipc-metadata-list-summary-item")

    for element in movie_elements:
        title = element.find_element(By.CSS_SELECTOR, "h3.ipc-title__text").text
        year = element.find_elements(By.CSS_SELECTOR, "span.sc-d5ea4b9d-7")[0].text
        runtime = element.find_elements(By.CSS_SELECTOR, "span.sc-d5ea4b9d-7")[1].text
        rating = element.find_element(By.CSS_SELECTOR, "span.ipc-rating-star--rating").text
        
        # Use try-except to handle potential missing element cases
        try:
            mpa_rating = element.find_elements(By.CSS_SELECTOR, "span.sc-d5ea4b9d-7")[2].text
        except IndexError:
            mpa_rating = "Not Rated"
        
        movies.append(Movie(title, year, runtime, rating, mpa_rating))

    # Go to the next page
    next_page_url = f"{base_url}&start={page * 50 + 1}&ref_=adv_nxt"
    driver.get(next_page_url)
    time.sleep(5)  # Wait for the next page to load

# Create a DataFrame
df = pd.DataFrame([vars(movie) for movie in movies])

# Remove the index numbers from the title column if they exist
df['title'] = df['title'].str.replace(r'^\d+\.\s+', '', regex=True)

# Print the DataFrame
print(df)

# Close the browser
driver.quit()
```
