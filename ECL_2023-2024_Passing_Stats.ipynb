```python
from selenium import webdriver 
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager 
import pandas as pd
```

the player table is loaded dynamically via JavaScript. requests and BeautifulSoup only retrieve the initial HTML source, not the content generated by JavaScript after the page loads.

```python
# Set up Chrome options
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run headless Chrome
```

```python
# Initialize WebDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)
```

```python
# URL of the page
url = 'https://fbref.com/en/comps/8/2023-2024/passing/2023-2024-Champions-League-Stats'
driver.get(url)
```

```python
# Let the page load
driver.implicitly_wait(10)  # seconds
```

```python
# Get the table by ID within the containing div
div = driver.find_element("id", "div_stats_passing")
table = div.find_element("id", "stats_passing")
```

```python
# Read the HTML table into a DataFrame
df = pd.read_html(table.get_attribute('outerHTML'))[0]

# Close the WebDriver
driver.quit()
```

```python
# Set pandas to display all columns 
pd.set_option('display.max_columns', None)
```

```python
df.columns = ['_'.join(col).strip() for col in df.columns.values]
```

```python
df
```

```python
# Clean and rename specific columns 
df.columns = [col.split('_')[-1].lower() if 'Unnamed' in col else col for col in df.columns]
```

```python
df
```

```python
# Strip and lower-case all column names 
df.columns = df.columns.str.strip().str.lower()

# Strip and lower-case the 'player' column values for matching 
df['player'] = df['player'].str.strip().str.lower()

# Identify rows where the 'player' column contains the value 'player' 
rows_to_drop = df[df['player'] == 'player'].index
rows_to_drop
```

```python
# Drop these rows 
df.drop(rows_to_drop, inplace=True) 

# Reset index 
df.reset_index(drop=True, inplace=True)
```

```python
# Select specified columns and sort by 'Ast' 
selected_columns = ['player', 'squad', 'pos', 'kp', 'ast', 'xag', 'expected_xa', 'expected_a-xag'] 
df_ast = df[selected_columns].sort_values(by='ast', ascending=False)
df_ast.head(30)
```

- **90s** Minutes divided by 90
- **PPA (Passes Per Attack):** This shows how many passes a player makes before attempting a shot on goal. It's a measure of their involvement in the build-up play.
- **CrsPA (Crosses Per Shot):** This indicates how many crosses a player makes for every shot they take. It can reflect their crossing ability and decision-making in the final third. Players with strong crossing ability often need to be fast and agile to get into the right positions to deliver crosses.
- **PrgP (Progressive Passes):** This tracks the number of passes a player makes that move the ball significantly towards the opponent's goal. It highlights their ability to advance the play.
- **1/3 (One-third):** This stat highlights how often a player or team is able to transition the ball into a more dangerous area, increasing the likelihood of creating scoring opportunities.

```python
# Select specified columns and sort by 'CrsPA' (Completed crosses into penalty area) 
cross_pass_columns = ['player', 'squad', 'pos','90s', 'ppa', 'crspa', 'prgp', '1/3'] 
df_cross = df[cross_pass_columns].sort_values(by='crspa', ascending=False)
df_cross.head(30)
```
